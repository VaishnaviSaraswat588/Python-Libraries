```python
import numpy as np
```

```python
np_array = np.array([1,2,3,4,5])
print(np_array)
print(type(np_array))
```

```python
# creating a 1 dim array
a = np.array([1,2,3,4])
print(a)
print(a.shape)

b = np.array([(1,2,3,4),(5,6,7,8)])
print(b)
print(b.shape)

c = np.array([(1,2,3,4),(5,6,7,8)], dtype= float)
print(c)
print(c.shape)
```

```python
# create an numpy array of zeros
x = np.zeros((4,5))
print(x)
```

```python
# create an numpy array of ones
y = np.ones((3,3))
print(y)
```

```python
# create an array of a particular value
z = np.full((5,4),5)
print(z)
```

```python
# create an identity matrix
a = np.eye(4)
print(a)
```

```python
# create an numpy array with random value
b = np.random.random((3,4))
print(b)
```

```python
# create an numpy with random integer values array within a specific range
c = np.random.randint(10,100,(3,5))
print(c)
```

```python
# array of evenly spaced values
d = np.linspace(10,30,5)
print(d)
# specifying the number of values required
```

```python
# array of evenly spaced values --> specifying the step
e = np.arange(10,30,5)
print(e)
```

```python
# Mathematical operations on an numpy array
```

```python
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]
print(list1+list2)      # concatenate or joins two list
```

```python
a = np.random.randint(0,10,(3,3))
b = np.random.randint(10,20,(3,3))
print(a)
print(b)
```

```python
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
a = np.random.randint(0,10,(3,3))
b = np.random.randint(10,20,(3,3))
print(a)
print(b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

```python
# array manipulation
```

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
# transpose
trans = np.transpose(array)
print(trans)
print(trans.shape)
```

```python
# another way of transpose
trans2 = array.T
print(trans2)
print(trans2.shape)
```

```python
# reshaping the array
a = np.random.randint(0,10,(2,3))
print(a)
print(a.shape)

b = a.reshape(3,2)
print(b)
```
